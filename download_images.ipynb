```python
import os
import shutil
from bing_image_downloader import downloader
```

```python
ROOT_DIR = os.getcwd()
# ROOT_DIR = "."

# QUERY_STRING = "leopard"
# QUERY_STRING = "zebra"

# QUERY_STRING = "sunflower"
# QUERY_STRING = "rose flower"
# QUERY_STRING = "lily flower"
# QUERY_STRING = "iris flower"
# QUERY_STRING = "pink tulip"

# QUERY_STRING = "Toyota Corolla"
QUERY_STRING = "Toyota Camry"
# QUERY_STRING = "Toyota RAV4"
# QUERY_STRING = "Ford Mustang"
# QUERY_STRING = "Ford Fiesta"
# QUERY_STRING = "Volkswagen Beetle"
# QUERY_STRING = "Honda Civic"
# QUERY_STRING = "Lamborghini Aventador"
# QUERY_STRING = "Jeep Wrangler"
# QUERY_STRING = "Chevrolet Silverado"

N_IMAGES = 150
```

```python
output_dir = os.path.join(ROOT_DIR, "dataset")

downloader.download(
    query=QUERY_STRING,
    limit=N_IMAGES,
    output_dir=output_dir,
    adult_filter_off=True,
    force_replace=False,
    timeout=60,
    filter="",
    verbose=True
)

images_dir = os.path.join(output_dir, QUERY_STRING)

shutil.make_archive(images_dir, 'zip', images_dir)
```
